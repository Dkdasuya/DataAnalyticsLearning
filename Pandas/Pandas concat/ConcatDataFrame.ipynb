Joining two dataframe using merge function

```python
india_temperature_dataframe = pd.DataFrame({
    'city':['mumbai','banglore','delhi'],
    'temperature':[21,24,30],
    
})
india_windspeed_dataframe = pd.DataFrame({
    'city':['banglore','delhi','mumbai'],
    'windspeed':[2,5,1],
    'humidity':[87,78,99]
})
```

```python
newdataframe = pd.merge(india_temperature_dataframe,india_windspeed_dataframe,on="city")
newdataframe    #It will only merge those rows which are common in both of the dataframes.
```

```python
india_temperature_dataframe = pd.DataFrame({
    'city':['mumbai','banglore','delhi','madhurai'],
    'temperature':[21,24,30,45],
    
})
india_windspeed_dataframe = pd.DataFrame({
    'city':['banglore','delhi','mumbai'],
    'windspeed':[2,5,1],
    'humidity':[87,78,99]
})
```

```python
newdataframe = pd.merge(india_temperature_dataframe,india_windspeed_dataframe,on="city",how="left",indicator=True)# The how can be different values 'left','outer','right' these are basically the joins 
newdataframe 
```

```python
india_temperature_dataframe = pd.DataFrame({
    'city':['mumbai','banglore','delhi','madhurai'],
    'temperature':[21,24,30,45],
    
})
suffixes_example = pd.merge(india_temperature_dataframe,india_temperature_dataframe,on='city',suffixes=['-left','-right'])
suffixes_example
```
